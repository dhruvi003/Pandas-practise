```python
# Which was the most-ordered item?
chipo.value_counts()
# select item_name
# from chipo 
# group by item_name 
# order by sum(quantity) desc
# limit 1 

a = chipo.groupby('item_name')['quantity'].sum()
a

```

```python
a.sort_values(ascending=False).head(1)
```

```python
# most ordered items value
chipo.columns
# 'order_id', 'quantity', 'item_name', 'choice_description','item_price'

chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(1)
```

```python
# How many items were orderd in total?
chipo.quantity.sum()
```

```python
# Turn the item price into a float
chipo.item_price.dtype

# <StringDtype(storage='python', na_value=nan)>
chipo.item_price = chipo.item_price.apply(lambda x : float(x[1:-1]))
```

```python
a = chipo.item_price[0]
a[1:-1]
```

```python
#  How much was the revenue for the period in the dataset?
# revenue = quantity * price 
r = (chipo['quantity'] * chipo['item_price']).sum()
r
```

```python
# How many orders were made in the period?
chipo.columns
chipo['order_id'].value_counts().count()
```

```python
# What is the average revenue amount per order?
chipo.columns
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
```

```python
chipo.columns
chipo.groupby('order_id')['revenue'].sum().mean()
```
